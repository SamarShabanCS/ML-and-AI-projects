```python
#reading data
path='csv/'
files=glob.glob("csv/*.csv")
data=[]
names=[]
for file in files:
    df=read_csv(file,header=None,names=['Av','Mx_rt','Mn_rt','Avg_rt','th'])
    #df['R_dates']=reading_dates
    ##df=df.set_index('R_dates')
    names.append(file)
    data.append(df)
```

```python
flatten_data=[]
for sublist in data:
    for item in sublist.values:
        flatten_data.append(item)
flatten_data=numpy.array(flatten_data)
flatten_data.shape
values_rt=flatten_data[:,1:4]
values_rt
```

```python
#decoded=Dense(2,activation='relu')(encoded)
decoded=Dense(3,activation='relu')(encoded)
```

```python
AE=Model(input_QoS,decoded)
```

```python
AE.compile(optimizer='adam',loss='mse')
```
